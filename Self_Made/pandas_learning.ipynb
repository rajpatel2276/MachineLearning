```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Data_Folder/titanic.csv")
```

```python
print(df.head())
print("-----------------------------------")
print(df.info())
print("-----------------------------------")
print(df.describe())
```

```python
df.columns
```

```python
df.columns = df.columns.str.strip()
df.columns = df.columns.str.lower()
```

```python
df = df.drop(columns=['cabin'])
```

```python
# Filling the numerical gaps
median_age = df['age'].median()
df['age'] = df['age'].fillna(median_age)
```

```python
# Droping missing values
df = df.dropna()
```

```python
print(f"Remaining rows:{len(df)}")
```

Feature engineering to get more insightfull informations

```python
df['familysize'] = df['sibsp'] + df['parch']
df = df.drop(columns=['sibsp','parch'])
```

Encoding Text to numbers

```python
# 1.Binary Encoding for the sex column
df['sex'] = df['sex'].map({'male':0,'female':1})
```

```python
# 2.One hot encoding for embarked column
df = pd.get_dummies(df,columns=['embarked'],drop_first=True)
```

Normalising the data to create a uniform calculation 

```python
cols_scale = ['age','fare','familysize']

for col in cols_scale:
    min_val = df[col].min()
    max_val = df[col].max()
    df[col] = (df[col] - min_val) / (max_val - min_val)
print(df.head())
```
