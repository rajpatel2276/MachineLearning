```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import matplotlib.pyplot as plt
```

```python
print(tf.__version__)
```

```python
X = np.array([[0.1], [0.2], [0.3], [0.6], [0.7], [0.8]], dtype=float)
y = np.array([[0], [0], [0], [1], [1], [1]], dtype=float)
```

```python
model = Sequential([
    Dense(4,activation='relu'),
    Dense(1,activation='sigmoid')
])
```

```python
#compiling the model
optimizer = tf.keras.optimizers.Adam(learning_rate=0.01)
model.compile(optimizer=optimizer,
             loss='binary_crossentropy',
             metrics=['accuracy'])
```

```python
history = model.fit(X,y , epochs=500,verbose=0)
```

```python
test_inputs = np.array([[0.9], [0.4]])
predictions = model.predict(test_inputs)
print("Predictions for [0.9, 0.4]:\n", predictions)
```

```python
# 6. Plot predictions across input range
x_range = np.linspace(0, 1, 100).reshape(-1, 1)
y_pred = model.predict(x_range)

plt.figure(figsize=(8,5))
plt.plot(x_range, y_pred, label="Predicted Probability")
plt.axhline(0.5, color='red', linestyle='--', label="Decision Boundary")
plt.scatter(X, y, color='black', label="Training Data")
plt.xlabel("Input Value")
plt.ylabel("Predicted Probability")
plt.title("Model Predictions Across Input Range")
plt.legend()
plt.show()

# 7. Plot loss curve
plt.figure(figsize=(8,5))
plt.plot(history.history['loss'], label="Training Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Loss Curve During Training")
plt.legend()
plt.show()
```
